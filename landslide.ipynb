```
import numpy as np
import pandas as pd
```

```
data = pd.read_csv("landslide.csv")
```

```
data.head()
```

```
data = data[['HAPPENED', 'MONTH', 'DAY', 'LATITUDE', 'LONGITUDE', 'DEATHS']]
data.head()
```

```
data = data.replace(to_replace= ['Yes','No'],value = ['1','0'])
```

```
data = data[data['HAPPENED'].notna()]
data = data[data['MONTH'].notna()]
data = data[data['DAY'].notna()]
data = data[data['LATITUDE'].notna()]
data = data[data['LONGITUDE'].notna()]
data = data[data['DEATHS'].notna()]
```

```
data.head()
```

```
from sklearn.model_selection import train_test_split

X=data.drop('HAPPENED', axis=1)  
y=data.HAPPENED  

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


from sklearn.ensemble import RandomForestRegressor

classifier = RandomForestRegressor(n_estimators = 1000)

classifier.fit(X_train, y_train);
```

```
y_pred = classifier.predict(X_train)
import sklearn.metrics as metrics
z = float(metrics.mean_squared_error(y_train, y_pred))
print(z)
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
import pickle
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
X = data.drop(['FATALITIES', 'INJURY_ESTIMATE'], axis=1) 
 
y = data['FATALITIES']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
classifier = RandomForestClassifier(n_estimators = 50, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
pred = classifier.predict(X_test)

accuracy = accuracy_score(y_test, pred)
print("accuracy: " + str(accuracy * 100) + "%")
```
